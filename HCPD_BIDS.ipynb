```python
import os 
import glob 
from bids_validator import BIDSValidator
from pathlib import Path
from tqdm import tqdm
```

```python
path = Path('/nfs2/harmonization/BIDS/HCPD/')
path.resolve()
val = BIDSValidator()
```

```python
#Passed the BIDS path to the BDS validator, else the validation will fail
bval = {}
for file in tqdm(path.rglob('*.bval')):
    name = file.name
    fil = file.parent.parent.name
    bids_path = '/' + str(fil) + '/'  +'dwi'+ '/'+ str(name)
    #print(name)
    #print('BIDS validation :', val.is_bids(bids_path))
    bval[bids_path] = val.is_bids(bids_path)
```
